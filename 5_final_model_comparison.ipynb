__Analyse final results__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import re
from time import time
```

```python
decept = pd .read_csv("data/models_decept.csv")
sentiment = pd .read_csv("data/models_sentiment.csv")
```

```python
decept.head()
```

```python
decept.sort_values(by=['Accuracy'], inplace=True, ascending=False)
```

```python
barlist=plt.barh(decept["Description"], decept["Accuracy"], color="darkgrey")
plt.title("model accuracy (deceptive opinion)")
barlist[0].set_color('r')
#plt.show()
plt.savefig('model_decept_accuracy.png')
```

```python
sentiment.sort_values(by=['Accuracy'], inplace=True, ascending=False)
```

```python
barlist=plt.barh(sentiment["Description"], sentiment["Accuracy"], color="darkgrey")
plt.title("model accuracy (sentiment)")
barlist[0].set_color('b')
#plt.show()
plt.savefig('model_sentiment_accuracy.png')
```
